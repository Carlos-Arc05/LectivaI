```python
#Leyendo datos ocupados
df_Ocupados = pd.read_csv("https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV",encoding='latin',sep=';')
```

```python
#Punto 1, 1.1 población total
Población_Total = df_cgsse['FEX_C18'].sum()
print("Población total nacionalestimada:", Población_Total)
```

```python
#Punto 1, 1.2 población en edad de trabajar
# Población en edad de trabajar (15 años o más)
#Calculo de la PET
PET = df_FuerzaTrabajo[df_FuerzaTrabajo['PET']==1]['FEX_C18'].sum()
print("Población Económicamente Trabajadora (PET):", PET)
```

```python
# Punto 1, 1.3 población no económicamente trabajadora
# Población No Económicamente Trabajadora (PNET)
PNET = Población_Total - PET
print("Población No Económicamente Trabajadora (PNET):" , PNET)
```

```python
# Punto 1, 1.4 Población economicamente activa
#Calculo PEA
PEA = df_FuerzaTrabajo[df_FuerzaTrabajo['FT']==1]['FEX_C18'].sum()
print("Población Económicamente Activa (PEA):", PEA)
#Punto 1, 1.5 Población economicamente inactiva
#Calculo PEI
PEI = df_FuerzaTrabajo[df_FuerzaTrabajo['FFT']==1]['FEX_C18'].sum()
print("Población Económicamente Inactiva (PEI):", PEI)
```

```python
#Punto 1, 1.6 Tasa de ocupación
Tasa_Ocupacion = (PEA / Población_Total) * 100
print("Tasa de Ocupación: {:.2f}%".format(Tasa_Ocupacion))
```

```python
#Punto 1, 1.6 Tasa de desempleo
Tasa_Desempleo = (PEI / PEA) * 100
print("Tasa de Desempleo: {:.2f}%".format(Tasa_Desempleo))
```

```python
#Punto 2 Tasa desempleo en hombres y mujeres en todos los departamentos, por lo cual se debe realizar un merge para obtener los datos de genero
df_no_ocupado_genero = pd.merge(
    df_no_ocupados[df_no_ocupados["DSI"]==1], 
    df_cgsse[['DIRECTORIO','SECUENCIA_P','ORDEN','P3271']], 
    on=['DIRECTORIO','SECUENCIA_P','ORDEN'], 
    how='left')
#Se agrupa por departamento y genero
desocupados_sexo = (df_no_ocupado_genero.groupby(['DPTO','P3271'])['FEX_C18'].sum().reset_index().rename(columns={'FEX_C18':'desocupados'}))

#Tambien agregamos a los ocupados para la medición
df_Ocupados_gen = pd.merge(
    df_Ocupados, df_cgsse[['DIRECTORIO','SECUENCIA_P','ORDEN','P3271']], 
    on=['DIRECTORIO','SECUENCIA_P','ORDEN'], 
    how='left')
Ocupados_sexo = (df_Ocupados_gen.groupby(['DPTO','P3271'])['FEX_C18'].sum().reset_index().rename(columns={'FEX_C18':'ocupados'}))

#Unimos ambos dataframes 
Pea_sexo = pd.merge(Ocupados_sexo, desocupados_sexo, on=['DPTO','P3271'], how='outer').fillna(0)
Pea_sexo['PEA'] = Pea_sexo['ocupados'] + Pea_sexo['desocupados']


#Tasa de desempleo (TD)
Pea_sexo['TD'] = Pea_sexo['desocupados'] / Pea_sexo['PEA'] * 100

#Reemplazamos sexo por etiquetas
Pea_sexo['Sexo'] = Pea_sexo['P3271'].map({1:'Hombres', 2:'Mujeres'})
# Hacemos el print
print(Pea_sexo[['DPTO','Sexo','TD']].head(20))
```

```python
#Punto 2.2, mayor tasa de desempleo dada por genero
td_max = Pea_sexo.loc[Pea_sexo.groupby("Sexo")["TD"].idxmax()]
td_min = Pea_sexo.loc[Pea_sexo.groupby("Sexo")["TD"].idxmin()]

print("Departamentos con MAYOR tasa de desempleo por genero:")
print(td_max[['DPTO','Sexo','TD']])

print("\nDepartamentos con MENOR tasa de desempleo por genero:")
print(td_min[['DPTO','Sexo','TD']])
```

```python
#Punto 3, se debe realizar un merge para obtener los datos (P3271: Hombres = 1, Mujeres = 2)
df_desocupados_genero = pd.merge(df_no_ocupados,df_cgsse[["DIRECTORIO", "SECUENCIA_P", "ORDEN", "P3271"]], on=["DIRECTORIO", "SECUENCIA_P", "ORDEN"], how="left")

# Calcular hombres desocupados
PHDO = df_desocupados_genero[df_desocupados_genero["P3271"] == 1]["FEX_C18"].sum()
print(f"Hombres desempleados: {int(PHDO)}")

# Calcular mujeres desocupadas
PMDO = df_desocupados_genero[df_desocupados_genero["P3271"] == 2]["FEX_C18"].sum()
print(f"Mujeres desempleadas: {int(PMDO)}")
```

```python
#Punto 3, parte dos, clasificamos los grupos de edad para cada departamento, tambien para idenficicar el grupo con mayor y menor tasa de desempleo
#Return se usara para determinar menores de 18 y mayores entre 50-59 que no clasifican en la muestra
def clasificar_edad(x):
    if 18 <= x < 24:
        return "18-24"
    elif 24 <= x < 30:
        return "24-30"
    elif 30 <= x < 40:
        return "30-40"
    elif 40 <= x < 50:
        return "40-50"
    elif x >= 60:
        return "60+"
    else:
        return "Otros"
```

```python
#Armamos la PEA por departamento y grupo de edad, hacemos merge para traer la edad (P6040) tanto en ocupados como en desocupados.
#Agrupamos por DPTO y grupo de edad.
#Calculamos TD como antes
# Creamos columnas para grupo de edad en df_cgsse
df_cgsse["grupo_edad"] = df_cgsse["P6040"].apply(clasificar_edad)

# Desocupados con edad
df_desocupados_gen = pd.merge(df_no_ocupados[df_no_ocupados["DSI"]==1], df_cgsse[["DIRECTORIO","SECUENCIA_P","ORDEN","grupo_edad"]], on=["DIRECTORIO","SECUENCIA_P","ORDEN"], how="left")

desocupados_edad = (df_desocupados_gen.groupby(["DPTO","grupo_edad"])["FEX_C18"].sum().reset_index().rename(columns={"FEX_C18":"desocupados"}))

# Ocupados con edad
df_ocupados_gen = pd.merge(df_Ocupados, df_cgsse[["DIRECTORIO","SECUENCIA_P","ORDEN","grupo_edad"]], on=["DIRECTORIO","SECUENCIA_P","ORDEN"], how="left")

ocupados_edad = (df_ocupados_gen.groupby(["DPTO","grupo_edad"])["FEX_C18"].sum().reset_index().rename(columns={"FEX_C18":"ocupados"}))

# PEA y TD
pea_edad = pd.merge(ocupados_edad, desocupados_edad, on=["DPTO","grupo_edad"], how="outer").fillna(0)
pea_edad["PEA"] = pea_edad["ocupados"] + pea_edad["desocupados"]
pea_edad["TD"] = pea_edad["desocupados"] / pea_edad["PEA"] * 100
```

```python
# extraemos la información relevante
pea_edad = pea_edad[pea_edad["grupo_edad"].isin(["18-24","24-30","30-40","40-50","60+"])] 

# Para cada departamento, el grupo de mayor y menor TD
mayores = pea_edad.loc[pea_edad.groupby("DPTO")["TD"].idxmax()]
menores = pea_edad.loc[pea_edad.groupby("DPTO")["TD"].idxmin()]

print("Grupo con Mayor tasa de desempleo por departamento:")
print(mayores[["DPTO","grupo_edad","TD"]])
```

```python
#En este caso se imprime el de menor tasa de desempleo
print("\nGrupo con menor tasa de desempleo por departamento:")
print(menores[["DPTO","grupo_edad","TD"]])
```

```python
#Para la tasa de desempleo se evidencia que entre los primeros 10 departamentos con mayor tasa de desempleo el 80% se ve reflejada en los jovenes entre 18 y 24 años, el otro 20% de los primeros 10 muestras son entre los 24 y 30 años, reflejando
#un panorama delicado para el futuro de las familias del país ya que es posible que muchos de esos jovenes tengan un rol importante para la manutención en sus hogares
```

```python
#Punto 4, empezaremos con filtrar mujeres de 18 a 24 años de edad
df_cgsse["grupo_edad"] = df_cgsse["P6040"].apply(clasificar_edad)
```

```python
# Añadimos la columna 'grupo_edad' a df_Ocupados_gen
if 'grupo_edad' not in df_Ocupados_gen.columns:
	df_Ocupados_gen = pd.merge(
		df_Ocupados_gen,
		df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'grupo_edad']],
		on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'],
		how='left'
	)

df_mujeres_oc = df_Ocupados_gen[(df_Ocupados_gen["P3271"]==2) & (df_Ocupados_gen["grupo_edad"]=="18-24")]

# Añadimos la columna 'grupo_edad' a df_no_ocupado_genero si no existe
if 'grupo_edad' not in df_no_ocupado_genero.columns:
	df_no_ocupado_genero = pd.merge(
		df_no_ocupado_genero,
		df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'grupo_edad']],
		on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'],
		how='left'
	)

df_mujeres_des = df_no_ocupado_genero[(df_no_ocupado_genero["P3271"]==2) & (df_no_ocupado_genero["grupo_edad"]=="18-24")]
```

```python
#Merge de ocupados y desocupados con sexo y edad
df_ocupados_gen = pd.merge(df_Ocupados, df_cgsse[["DIRECTORIO","SECUENCIA_P","ORDEN","P3271","grupo_edad"]], on=["DIRECTORIO","SECUENCIA_P","ORDEN"], how="left")
df_desocupados_gen = pd.merge(df_no_ocupados[df_no_ocupados["DSI"]==1], df_cgsse[["DIRECTORIO","SECUENCIA_P","ORDEN","P3271","grupo_edad"]], on=["DIRECTORIO","SECUENCIA_P","ORDEN"], how="left")
```

```python
#Agrupamos por departamento la información
ocupadas = df_mujeres_oc.groupby("DPTO")["FEX_C18"].sum().reset_index().rename(columns={"FEX_C18":"ocupadas"})
desocupadas = df_mujeres_des.groupby("DPTO")["FEX_C18"].sum().reset_index().rename(columns={"FEX_C18":"desocupadas"})
```

```python
#Calcular PEA (Población economicamente activa) y TD (Tasa de desempleo)
pea = pd.merge(ocupadas, desocupadas, on="DPTO", how="outer").fillna(0)
pea["PEA"] = pea["ocupadas"] + pea["desocupadas"]
pea["TD"] = pea["desocupadas"] / pea["PEA"] * 100
```

```python
#Ordenar y mostrar los 3 departamentos con mayor TD
td_mujeres_18_24_min = pea.sort_values("TD").head(3)
print(td_mujeres_18_24_min)
#Los departamento del pais con menor tasa de desempleo en mujeres entre 18 y 24 años son: Amazónas, Vaupés y Quindío
```

```python
#Punto 5, calculo promedio de ingresos laborales Risaralda, Caldas, Qundío, Valle del cauca y Antioquia
# Filtrar ocupados en los departamentos a analizar
deptos_interes = [66, 17, 63, 5, 76]  # 66=Risaralda, 17=Caldas, 63=Quindío, 5=Antioquia, 76=Valle del Cauca
df_ingresos = df_Ocupados[df_Ocupados["DPTO"].isin(deptos_interes)]

# Calcular el promedio ponderado de ingresos laborales (INGLABO)
promedios_ingresos = (df_ingresos.groupby("DPTO").apply(lambda x: (x["INGLABO"] * x["FEX_C18"]).sum() / x["FEX_C18"].sum()).reset_index(name="Ingreso_promedio"))

# Reemplazar códigos de departamentos por nombres
mapa_dptos = {66:"Risaralda", 17:"Caldas", 63:"Quindío", 5:"Antioquia", 76:"Valle del Cauca"}
promedios_ingresos["Departamento"] = promedios_ingresos["DPTO"].map(mapa_dptos)

# Reordenar columnas
promedios_ingresos = promedios_ingresos[["Departamento","Ingreso_promedio"]]

print(promedios_ingresos)

#El departamento con mayor ingreso promedio es Antioquia, seguido de Caldas, el Valle del Cauca se situa en el ultimo lugar.
```
